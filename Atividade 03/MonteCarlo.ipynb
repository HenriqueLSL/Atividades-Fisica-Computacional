# Monte Carlo 


```python
import numpy as np
'''
Calcula o valor de pi  através de uma aproximação de Monte Carlo.

'''
def pi_monte_carlo(n):
    num = np.random.uniform(0, 1, size=(n, 2))
    x = num[:, 0]
    y = num[:, 1]
    j = 0
    for i in range(0, n):
        r = (x[i]**2 + (y[i])**2)**0.5
        if r<= 1:
            j += 1
    return j, n
p = pi_monte_carlo(100000)[0]/pi_monte_carlo(100000)[1]
print(4*p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
''' 
Estima o valor de pi usando a simulação de Monte Carlo.

Gera pontos aleatórios em um quadrado de lado 1 e conta quantos
caem dentro de um círculo inscrito nesse quadrado. A razão entre
o número de pontos dentro do círculo e o número total de pontos
é usada para aproximar a razão entre as áreas do círculo e do
quadrado, permitindo estimar o valor de pi.

'''
def salvar_pontos(lista, pontos):
  lista += pontos

def pi_monte_carlo(n, retorna_pontos=False):
  salvar = lambda lista, pontos: None
  if retorna_pontos == True:
    salvar = salvar_pontos

  total = n
  num = np.random.uniform(0, 1, size=(total, 2))
  x = num[:, 0]
  y = num[:, 1]
  j = 0
  dentro = []
  fora = []
  for i in range(0, total):
      r = ((x[i])**2 + (y[i])**2)**0.5
      pontos = (x[i], y[i])
      
      if r <= 1:
          j += 1
          salvar(dentro, [pontos])
      else:
          salvar(fora, [pontos])

  pi_aproximado = 4 * (j / total)
  return pi_aproximado, dentro, fora
 
pi_aproximado, dentro, fora = pi_monte_carlo(10000, retorna_pontos=True)
print(pi_aproximado)
plt.scatter(*zip(*dentro), color='blue', s=1) 
plt.scatter(*zip(*fora), color='red', s=1)   
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pi_monte_carlo(N):
  num = np.random.uniform(0, 1, size=(N, 2))
  x = num[:,0]
  y = num[:,1]
  j = 0
  for i in range(0, N):
    r = ((x[i])**2 + (y[i])**2)**0.5
    if r <= 1:
      j += 1
  P = j/N
  pi = 4*P
  return pi

En = []
N = []
p = 10

for i in range(1, 7):
  N = N + [p]
  cont = 0
  for j in range(1, 10):
    Erro = ((abs(np.pi - pi_monte_carlo(p)))/np.pi)
    cont += Erro
  p *= 10
  media = cont/10
  En.append(media)
```

```python
plt.subplot(221)
plt.plot(N, En, 'o')

plt.subplot(222)
plt.loglog(N, En, 'o')

plt.subplot(223)
plt.semilogx(N, En, 'o')

plt.subplot(224)
plt.semilogy(N, En, 'o')

plt.tight_layout()
plt.show()
```

```python
plt.loglog(N, En, 'o')

x1 = np.log10(N)
y1 = np.log10(En)
param = np.polyfit(x1, y1, 1) # y1 = a*x1 + b
print(f'a = {param[0]}, b = {param[1]}')
# y1 = a*x1 + b
# ln(En) = a * ln(N) + b
# En = e^b * N^a


x_fit = np.linspace(1, 6, 5)
y_fit = np.polyval(param, x_fit)
plt.plot(10**(x_fit), 10**(y_fit),'r--', ms=1)

plt.show()
```
