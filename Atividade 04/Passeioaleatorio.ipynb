# Passeio Aleatório 

```python
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
  """
  Simula um passeio aleatório unidimensional.

  """
  rng = np.random.default_rng()
  posicao = 0
  posicoes = [posicao]
  for _ in range(N):
    if rng.random() < p:
      posicao += 1
    else:
      posicao -= 1
    posicoes.append(posicao)
  return np.array(posicoes)


N = 1000
posicoes_p05 = passeio_aleatorio(N, 0.5)
posicoes_p08 = passeio_aleatorio(N, 0.8)

plt.plot(posicoes_p05, label='p = 0.5')
plt.plot(posicoes_p08, label='p = 0.8')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
  """
  Simula um passeio aleatório unidimensional.

  """
  rng = np.random.default_rng()
  posicao = 0
  posicoes = [posicao]
  for _ in range(N):
    if rng.random() < p:
      posicao += 1
    else:
      posicao -= 1
    posicoes.append(posicao)
  return np.array(posicoes)

def calcular_media_desvio_padrao(N, p, num_passeios):
  """
  Calcula a média e o desvio padrão da posição do caminhante
  em função do número de passos.
  """
  medias = []
  desvios_padrao = []
  for i in range(1, N + 1):
    posicoes_finais = []
    for _ in range(num_passeios):
      posicoes = passeio_aleatorio(i, p)
      posicoes_finais.append(posicoes[-1])
    medias.append(np.mean(posicoes_finais))
    desvios_padrao.append(np.std(posicoes_finais))
  return np.array(medias), np.array(desvios_padrao)

N = 1000
num_passeios = 100


medias_p05, desvios_p05 = calcular_media_desvio_padrao(N, 0.5, num_passeios)
medias_p06, desvios_p06 = calcular_media_desvio_padrao(N, 0.6, num_passeios)


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(medias_p05, label='Média (p=0.5)')
plt.plot(desvios_p05, label='Desvio Padrão (p=0.5)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(medias_p06, label='Média (p=0.6)')
plt.plot(desvios_p06, label='Desvio Padrão (p=0.6)')
plt.legend()

plt.tight_layout()
plt.show()
```
